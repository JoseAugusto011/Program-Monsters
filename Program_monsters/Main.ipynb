```python
from math import sqrt
def distancia(xA,xB,yA,yB):
    distAB = sqrt((xA-xB)**2) + ((yA-yB)**2)
    return distAB
```

```python
def soma_posicao(xA,xB,yA,yB):
    
    nova_posicao=(xA+yA,xB+yB)
    return nova_posicao
```

```python


def movimento_monstrinho (ponto_monstro,ponto_central):
    
    
    
    lista_possibilidades = [] #lista com os locais que podem ir
    
    
   #restrição de locais
    if (convert_rgba4hex (data_mapa[ponto_monstro[1]//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis) and ( convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis):
       b = 0
    else:
        lista_possibilidades.append((-1,0))      


    
    if ( convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,ponto_monstro[0]//2]) in cores_inpassaveis) and ( convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis ):
        b = 0
    else:
    
        lista_possibilidades.append((0,-1))

    
    
    if  ( convert_rgba4hex (data_mapa[(ponto_monstro[1])//2,ponto_monstro[0]//2]) in cores_inpassaveis) and (convert_rgba4hex (data_mapa[ponto_monstro[1]//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis):
        b=0
    else:      
        lista_possibilidades.append((0,1))   
     
    
    if  (convert_rgba4hex (data_mapa[ponto_monstro[1]//2,(ponto_monstro[0])//2]) in cores_inpassaveis) and (convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,ponto_monstro[0]//2]) in cores_inpassaveis):
        b = 0
    else:
        lista_possibilidades.append((1,0))
        
        
    #verifica se pode se movimentar
    diferenca_x = ponto_central[0]-ponto_monstro[0]
    diferenca_y = ponto_central[1]-ponto_monstro[1] 
    
    if (diferenca_x < 3 and diferenca_x >-3) or (diferenca_y < 3 and diferenca_y >-3):      
        
        
        
        if (diferenca_x<0) and ((-1,0) in lista_possibilidades):
            ponto_monstro[0]-=1
        elif (diferenca_x>0) and (1,0) in lista_possibilidades:
            ponto_monstro[0]+=1        
        elif (diferenca_y<0)  and (0,-1) in lista_possibilidades:
            ponto_monstro[1]-=1
        elif (diferenca_y>0) and (0,1) in lista_possibilidades:
            ponto_monstro[1]+=1
    else:
            
        possivel = random.choice(lista_possibilidades)   
        ponto_monstro[0]+=possivel[0]
        ponto_monstro[1]+=possivel[1]  
            
        
        
    return ponto_monstro
```

### Calcula a vez do personagem, tendo preferência o personagem de maior velocidade

```python
def calcular_vez():
    
   
    if (status0['habilidades'][3] ==  status1['habilidades'][3]):
        vezes = [0,1]
    
    else:
        
        vezes = []
        limite = status0['habilidades'][3] +  status1['habilidades'][3]
        vez1,vez2 = status0['habilidades'][3], status1['habilidades'][3]
        
        while (len(vezes) < 25) or (vez1+vez2 != 0):
        
            vez1 +=  status0['habilidades'][3]
            vez2 +=  status1['habilidades'][3]
            
            if vez1 >= limite:
                vez1-=limite
                vezes.append(0)
                
            if vez2 >= limite:
                vez2-=limite
                vezes.append(1)
        
    return vezes
```

```python
def movimento_monstrinho (ponto_monstro,ponto_central):
    
    
    
    lista_possibilidades = [] #lista com os locais que podem ir
    
    
   #restrição de locais
    if (convert_rgba4hex (data_mapa[ponto_monstro[1]//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis) and ( convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis):
       b = 0
    else:
        lista_possibilidades.append((-1,0))      


    
    if ( convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,ponto_monstro[0]//2]) in cores_inpassaveis) and ( convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis ):
        b = 0
    else:
    
        lista_possibilidades.append((0,-1))

    
    
    if  ( convert_rgba4hex (data_mapa[(ponto_monstro[1])//2,ponto_monstro[0]//2]) in cores_inpassaveis) and (convert_rgba4hex (data_mapa[ponto_monstro[1]//2,(ponto_monstro[0]-1)//2]) in cores_inpassaveis):
        b=0
    else:      
        lista_possibilidades.append((0,1))   
     
    
    if  (convert_rgba4hex (data_mapa[ponto_monstro[1]//2,(ponto_monstro[0])//2]) in cores_inpassaveis) and (convert_rgba4hex (data_mapa[(ponto_monstro[1]-1)//2,ponto_monstro[0]//2]) in cores_inpassaveis):
        b = 0
    else:
        lista_possibilidades.append((1,0))
        
        
    #verifica se pode se movimentar
    diferenca_x = ponto_central[0]-ponto_monstro[0]
    diferenca_y = ponto_central[1]-ponto_monstro[1] 
    
    if (diferenca_x < 3 and diferenca_x >-3) or (diferenca_y < 3 and diferenca_y >-3):      
        
        
        
        if (diferenca_x<0) and ((-1,0) in lista_possibilidades):
            ponto_monstro[0]-=1
        elif (diferenca_x>0) and (1,0) in lista_possibilidades:
            ponto_monstro[0]+=1        
        elif (diferenca_y<0)  and (0,-1) in lista_possibilidades:
            ponto_monstro[1]-=1
        elif (diferenca_y>0) and (0,1) in lista_possibilidades:
            ponto_monstro[1]+=1
    else:
            
        possivel = random.choice(lista_possibilidades)   
        ponto_monstro[0]+=possivel[0]
        ponto_monstro[1]+=possivel[1]  
            
        
        
    return ponto_monstro
```
